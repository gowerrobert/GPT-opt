```python
%load_ext autoreload
%autoreload 2

from plot import load_outputs, plot_final_loss_vs_lr, plot_tuned_curves
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "serif"
plt.rcParams['font.size'] = 12
plt.rcParams['axes.linewidth'] = 1.5
plt.rc('text', usetex=False)
plt.rc('mathtext', fontset='cm')   # optional: use Computer Modern fontsplt.rc('legend', fontsize=10)
```

```python
output_dir = "gptopt/outputs/finewebmini_best/"
outputs = load_outputs(output_dir)

# 2) extract unique method names (before the "-lr-" suffix)
methods = sorted({ output['name'].split('-lr-')[0] for output in outputs })

# Sort outputs by the order of methods
outputs = sorted(outputs, key=lambda x: methods.index(x['name'].split('-lr-')[0]))

# 3) get a distinct qualitative colormap (tab20, Set3, or seaborn color_palette)
# Option A: matplotlib's tab20 (20 distinct colors)
base_colors = plt.get_cmap("tab20").colors  

# Option B: seaborn "hls" or "husl" palette (works for any number of categories)
# base_colors = sns.color_palette("husl", len(methods))

# 4) assign each method a color
colormap = {m: base_colors[i % len(base_colors)] for i, m in enumerate(methods)}

# 5) plot
plot_final_loss_vs_lr(outputs, colormap, outfilename="dap-finewebmini-best", val=True)
plt.show()
```
